```python
df['host_name'] = df['host'].apply(lambda x: x['host_name'])
df['host_id'] = df['host'].apply(lambda x: x['host_id'])
df['host_response_time']=df['host'].apply(lambda i:pd.Series([i.get('host_response_time',None)]))


df["street"]=df["address"].apply(lambda x: x["street"])
df["location"]=df["address"].apply(lambda x: x["location"])
df["country"]=df["address"].apply(lambda x: x["country"])
df["loaction_type"]=df["location"].apply(lambda x:x["type"])
df["location_coordinates"]=df["location"].apply(lambda x:x["coordinates"])

df["latitude"]=df["location_coordinates"].apply(lambda x:x[0])
df["longitude"]=df["location_coordinates"].apply(lambda x:x[1])
df["availability_30"]=df["availability"].apply (lambda x:x["availability_30"])
df["availability_60"]=df["availability"].apply (lambda x:x["availability_60"])
df["availability_90"]=df["availability"].apply (lambda x:x["availability_90"])
df["availability_365"]=df["availability"].apply (lambda x:x["availability_365"])
df['average_rating'] = df['review_scores'].apply(lambda x: np.mean([x.get('review_scores_accuracy', 0), x.get('review_scores_cleanliness', 0), x.get('review_scores_checkin', 0), x.get('review_scores_communication', 0), x.get('review_scores_location', 0), x.get('review_scores_value', 0)]))
```

```python
df.drop("listing_url",axis=1,inplace=True)
df.drop("summary",axis=1,inplace=True)
df.drop("space",axis=1,inplace=True)
df.drop("description",axis=1,inplace=True)
df.drop("notes",axis=1,inplace=True)
df.drop("transit",axis=1,inplace=True)
df.drop("access",axis=1,inplace=True)
df.drop("interaction",axis=1,inplace=True)
df.drop("reviews_per_month",axis=1,inplace=True)
df.drop("last_scraped",axis=1,inplace=True)
df.drop("calendar_last_scraped",axis=1,inplace=True)
df.drop("first_review",axis=1,inplace=True)
df.drop("last_review",axis=1,inplace=True)
df.drop("extra_people",axis=1,inplace=True)
df.drop("guests_included",axis=1,inplace=True)
df.drop("number_of_reviews",axis=1,inplace=True)
df.drop("weekly_price",axis=1,inplace=True)
df.drop("monthly_price",axis=1,inplace=True)
df.drop("availability",axis=1,inplace=True)
df.drop("host",axis=1,inplace=True)
df.drop("images",axis=1,inplace=True)
df.drop("address",axis=1,inplace=True)
df.drop("review_scores",axis=1,inplace=True)
df.drop("street",axis=1,inplace=True)
df.drop("reviews",axis=1,inplace=True)

```

```python
df['bedrooms'].fillna(0, inplace=True)
df['beds'].fillna(0, inplace=True)
df['cleaning_fee'].fillna('Not Specified', inplace=True)
df['security_deposit'].fillna('Not Specified',inplace = True)
df['host_response_time'].fillna('Null',inplace = True)
df['average_rating']=df['average_rating'].replace(0,'Unrated')
```
